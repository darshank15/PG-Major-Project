```python
import os
import json
import nltk
import operator
from nltk.corpus import wordnet
# nltk.download('wordnet')
```

```python
files = []
for r, d, f in os.walk("./data_final"):
    for file in f:
        if '.json' in file:
            files.append(os.path.join(r, file))
```

```python
hash_count = {}
for f in files:
	with open(f) as json_file:
		post_data = json.load(json_file)
		for post_id in post_data.keys():
			text = post_data[post_id]["text_des"]
			text = text.lower()
			words = text.split(" ")
			local_hash_count = set()
			for word in words:
				if word == "":
					continue
				if word[0]=='#':
					if wordnet.synsets(word[1:]):
						local_hash_count.add(word)

			for item in local_hash_count:
				if item not in hash_count:
					hash_count[item] = 0
				hash_count[item]=hash_count[item]+1
```

```python
# print(hash_count)
sorted_x = sorted(hash_count.items(), key=operator.itemgetter(1), reverse=True)
```

```python
count=0
top_hashtags = []
for i in sorted_x:
    if count>=100:
        break
    count=count+1

    top_hashtags.append([i[0],i[1]])

# print(top_hashtags)
for i in top_hashtags:
    print(i)
#     print(i[0]," ",i[1])
```
